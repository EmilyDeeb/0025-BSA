**Q2 — Categorisation with CASE (10 marks)**

- Classify each census block by population size:
  - 'LARGE' if popn_total > 5000
  - 'MED' if popn_total > 2000
  - 'SMALL' otherwise

Return:
geoid
boroname
popn_total
category
Limit 20.

```python
con.sql("""
SELECT BORONAME, POPN_TOTAL, BLKID,
  CASE
    WHEN POPN_TOTAL > 5000 THEN 'LARGE'
    WHEN POPN_TOTAL > 2000 THEN 'MED'
    ELSE 'SMALL'
END AS CAT
FROM nyc_census_blocks
LIMIT 20;
""")
```

**Q3 — Spatial join: block → neighborhood (15 marks)**

- Assign each census block to a neighborhood polygon.

Return: geoid, neigh_name, boroname, Use an appropriate spatial predicate.

(Hint: blocks are polygons, neighborhoods are polygons.)

```python
con.sql("""
SELECT n.name, b.BORONAME, b.POPN_TOTAL
FROM nyc_neighborhoods AS n
JOIN nyc_census_blocks AS b
ON ST_contains (n.geom, b.geom);
""").show()
```

```python
con.sql("""
SELECT n.name, b.BORONAME, sum(b.POPN_TOTAL), round(sum(b.popn_black)/sum(b.popn_total)*100) AS PCT_BLACK
FROM nyc_neighborhoods AS n
JOIN nyc_census_blocks AS b
ON ST_contains (n.geom, b.geom)
WHERE b.POPN_TOTAL > 1
GROUP BY n.name, b.boroname
ORDER BY PCT_BLACK DESC
;
""").show()
```

**Q4 — Accessibility: blocks within 500m of subway (15 marks)**

For each borough, compute:

- number of census blocks that are within 500 meters of any subway station

Return:
boroname,
n_blocks_near_subway,
🧠 You must avoid double-counting blocks that are near multiple stations.

```python
con.sql("""
SELECT DISTINCT s.name, sum(b.POPN_TOTAL), count(b.BLKID)
FROM nyc_subway_stations AS s
JOIN nyc_census_blocks AS b
ON ST_Dwithin (s.geom, b.geom, 500)
GROUP BY s.name, b.BLKID
ORDER BY sum(b.popn_total) DESC
LIMIT 20
""").show()
```

**Q5 — “Worst-served neighborhoods” (20 marks)**

Find the top 10 neighborhoods with the lowest subway station density defined as:

`\text{stations per km}^2 = \frac{\text{# stations in neighborhood}}{\text{area in km}^2}`

Return:
- neigh_name
- boroname
- n_stations
- area_km2
- stations_per_km2
- Order by stations_per_km2 ASC.
Handle neighborhoods with zero stations correctly (they should appear).

```python
con.sql("""
SELECT
  n.name,
  n.BORONAME,
  ROUND(ST_AREA(n.geom)) AS area_m2,
  COUNT (s.name) AS station_count,
  ROUND(
      COUNT(s.name) / (ST_AREA(n.geom) / 100000),
      2
    ) AS stations_per_km2
FROM nyc_neighborhoods AS n
JOIN nyc_subway_stations AS s
  ON ST_contains (n.geom, s.geom)
GROUP BY n.name, n.BORONAME, n.geom
ORDER by stations_per_km2 ASC
LIMIT 10;
""").show()
```
